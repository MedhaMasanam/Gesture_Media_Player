```python
import cv2
import vlc
import time
import os
import sys
import mediapipe as mp
import numpy as np
from collections import deque

# Paths (verify these exist)
face_cascade_path = r"C:\Users\Dhanu\gesture_control\haarcascade_frontalface_default.xml"
video_path = r"sample.mp4"
vlc_path = r"C:\Program Files\VideoLAN\VLC\vlc.exe"

# Verify paths
if not os.path.exists(face_cascade_path):
    print(f"Error: Face cascade file not found at {face_cascade_path}")
    sys.exit(1)
if not os.path.exists(video_path):
    print(f"Error: Video file not found at {video_path}")
    sys.exit(1)
if not os.path.exists(vlc_path):
    print(f"Warning: VLC executable not found at {vlc_path}. Ensure VLC is installed.")

# Initialize VLC Player
try:
    Instance = vlc.Instance(["--no-xlib", "--quiet"])
    player = Instance.media_player_new()
    if not player:
        raise RuntimeError("Could not create media player instance")
    
    Media = Instance.media_new(video_path)
    if not Media:
        raise RuntimeError("Could not create media instance")
    
    player.set_media(Media)
except Exception as e:
    print(f"VLC initialization failed: {str(e)}")
    sys.exit(1)

# Initialize Face Detector
face_cascade = cv2.CascadeClassifier(face_cascade_path)
if face_cascade.empty():
    print("Error: Could not load face cascade")
    sys.exit(1)

# Initialize MediaPipe Hands
mp_hands = mp.solutions.hands
hands = mp_hands.Hands(max_num_hands=1, min_detection_confidence=0.6, min_tracking_confidence=0.6)
mp_draw = mp.solutions.drawing_utils

# Initialize Webcam
cap = cv2.VideoCapture(0)
if not cap.isOpened():
    print("Error: Could not open webcam. Trying camera index 1...")
    cap = cv2.VideoCapture(1)
    if not cap.isOpened():
        print("Error: Could not open any webcam")
        sys.exit(1)

# Set webcam resolution
cap.set(cv2.CAP_PROP_FRAME_WIDTH, 640)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 480)

# Create a window for the webcam feed
cv2.namedWindow('Face and Gesture Control', cv2.WINDOW_NORMAL)

# Playback and gesture states
is_playing = False
last_gesture = None
gesture_cooldown = 0.6  # Balanced for responsiveness
last_gesture_time = 0
finger_counts = deque(maxlen=5)  # Store last 5 finger counts
consistent_frames = 0
last_finger_count = 0

try:
    # Start playback
    if player.play() == -1:
        print("Error: Could not start playback")
        sys.exit(1)
    
    time.sleep(0.5)  # Wait for VLC to initialize
    
    while True:
        ret, frame = cap.read()
        if not ret:
            print("Error: Could not read frame from webcam")
            break
            
        # Flip frame for natural view
        frame = cv2.flip(frame, 1)
        
        # Face detection
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        faces = face_cascade.detectMultiScale(gray, scaleFactor=1.05, minNeighbors=3, minSize=(30, 30))
        
        # Draw face rectangles
        for (x, y, w, h) in faces:
            cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2)
        
        # Print face detection status
        print(f"Detected {len(faces)} faces")

        # Hand gesture detection
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        results = hands.process(frame_rgb)
        
        gesture = None
        raised_fingers = 0
        if results.multi_hand_landmarks and len(faces) > 0:  # Only process gestures if face is detected
            for hand_landmarks in results.multi_hand_landmarks:
                # Draw hand landmarks and connections
                mp_draw.draw_landmarks(frame, hand_landmarks, mp_hands.HAND_CONNECTIONS)
                
                # Get landmark positions
                landmarks = hand_landmarks.landmark
                wrist = landmarks[0]  # Wrist
                palm_base = landmarks[5]  # Base of index finger
                index_tip = landmarks[8]  # Index finger tip
                middle_tip = landmarks[12]  # Middle finger tip
                ring_tip = landmarks[16]  # Ring finger tip
                pinky_tip = landmarks[20]  # Pinky finger tip
                
                # Calculate hand size
                hand_size = ((wrist.x - palm_base.x)**2 + (wrist.y - palm_base.y)**2)**0.5
                if hand_size < 0.1:  # Hand too small
                    print("Hand too small, ignoring")
                    continue
                
                # Dynamic thresholds
                y_threshold = hand_size * 0.45  # Slightly stricter
                x_separation = hand_size * 0.1  # Relaxed for natural spacing
                
                # Count raised fingers
                finger_tips = [index_tip, middle_tip, ring_tip, pinky_tip]
                raised_fingers = 0
                for i, tip in enumerate(finger_tips):
                    tip_to_palm_dist = ((tip.x - palm_base.x)**2 + (tip.y - palm_base.y)**2)**0.5
                    if (tip.y < palm_base.y - y_threshold and
                        abs(tip.x - palm_base.x) > x_separation and
                        tip_to_palm_dist > hand_size * 0.4):
                        raised_fingers += 1
                        # Draw green circle and thicker line on detected finger tip
                        h, w, _ = frame.shape
                        tip_pos = (int(tip.x * w), int(tip.y * h))
                        cv2.circle(frame, tip_pos, 5, (0, 255, 0), -1)
                        cv2.line(frame, (int(wrist.x * w), int(wrist.y * h)), tip_pos, (0, 255, 0), 3)
                    else:
                        # Draw red circle on undetected finger tip
                        h, w, _ = frame.shape
                        tip_pos = (int(tip.x * w), int(tip.y * h))
                        cv2.circle(frame, tip_pos, 5, (0, 0, 255), -1)
                
                # Stabilize finger count
                finger_counts.append(raised_fingers)
                if len(finger_counts) == finger_counts.maxlen:
                    # Simple majority voting
                    raised_fingers = max(set(finger_counts), key=list(finger_counts).count)
                    confidence = list(finger_counts).count(raised_fingers) / len(finger_counts)
                
                # Require consistent frames
                if raised_fingers == last_finger_count:
                    consistent_frames += 1
                else:
                    consistent_frames = 0
                    last_finger_count = raised_fingers
                
                # Determine gesture
                current_time = time.time()
                if (current_time - last_gesture_time > gesture_cooldown and
                    consistent_frames >= 3 and
                    confidence > 0.6):
                    if raised_fingers == 1:
                        gesture = "one_finger"
                    elif raised_fingers == 2:
                        gesture = "two_fingers"
                    elif raised_fingers == 3:
                        gesture = "three_fingers"
                    elif raised_fingers == 4:
                        gesture = "four_fingers"
                
                # Debug hand detection
                print(f"Hand detected. Raised fingers: {raised_fingers}, Confidence: {confidence:.2f}, "
                      f"Consistent frames: {consistent_frames}, Gesture: {gesture or 'None'}")
        
        # Save debug frame if no gesture after multiple attempts
        if not gesture and raised_fingers > 0 and consistent_frames < 3:
            cv2.imwrite(f"debug_gesture_{int(time.time())}.jpg", frame)

        # Control playback based on face detection
        if len(faces) > 0:
            if player.get_state() != vlc.State.Playing:
                player.play()
                is_playing = True
                print("Playing ▶️ (Face Detected)")
        else:
            if player.get_state() == vlc.State.Playing:
                player.pause()
                is_playing = False
                print("Paused ⏸️ (No Face)")

        # Control volume and seeking based on gestures
        if gesture and gesture != last_gesture and current_time - last_gesture_time > gesture_cooldown:
            if gesture == "one_finger":
                current_volume = player.audio_get_volume()
                new_volume = min(current_volume + 10, 100)
                player.audio_set_volume(new_volume)
                print(f"Volume Up 📈: {new_volume}%")
            elif gesture == "two_fingers":
                current_volume = player.audio_get_volume()
                new_volume = max(current_volume - 10, 0)
                player.audio_set_volume(new_volume)
                print(f"Volume Down 📉: {new_volume}%")
            elif gesture == "three_fingers":
                current_time_ms = player.get_time()
                player.set_time(current_time_ms + 3000)  # Forward 3 seconds
                print("Fast Forward 3s ⏩")
            elif gesture == "four_fingers":
                current_time_ms = player.get_time()
                player.set_time(max(current_time_ms - 3000, 0))  # Backward 3 seconds
                print("Rewind 3s ⏪")
            
            last_gesture_time = current_time

        last_gesture = gesture

        # Display status on frame
        status = "Playing" if is_playing else "Paused"
        cv2.putText(frame, f"Status: {status}", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
        cv2.putText(frame, f"Gesture: {gesture or 'None'}", (10, 60), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
        cv2.putText(frame, f"Fingers: {raised_fingers}", (10, 90), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

        cv2.imshow('Face and Gesture Control', frame)

        # Press 'q' to quit
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

finally:
    cap.release()
    cv2.destroyAllWindows()
    player.stop()
    hands.close()
```

```python
import cv2
import vlc
import time
import os
import sys
import mediapipe as mp
import numpy as np
from collections import deque

# Paths (verify these exist)
face_cascade_path = r"C:\Users\Dhanu\gesture_control\haarcascade_frontalface_default.xml"
video_path = r"C:\Users\Dhanu\Downloads\JISOO - earthquake (Official Music Video).mp4"
vlc_path = r"C:\Program Files\VideoLAN\VLC\vlc.exe"

# Verify paths
if not os.path.exists(face_cascade_path):
    print(f"Error: Face cascade file not found at {face_cascade_path}")
    sys.exit(1)
if not os.path.exists(video_path):
    print(f"Error: Video file not found at {video_path}")
    sys.exit(1)
if not os.path.exists(vlc_path):
    print(f"Warning: VLC executable not found at {vlc_path}. Ensure VLC is installed.")

# Initialize VLC Player
try:
    Instance = vlc.Instance(["--no-xlib", "--quiet"])
    player = Instance.media_player_new()
    if not player:
        raise RuntimeError("Could not create media player instance")
    
    Media = Instance.media_new(video_path)
    if not Media:
        raise RuntimeError("Could not create media instance")
    
    player.set_media(Media)
except Exception as e:
    print(f"VLC initialization failed: {str(e)}")
    sys.exit(1)

# Initialize Face Detector
face_cascade = cv2.CascadeClassifier(face_cascade_path)
if face_cascade.empty():
    print("Error: Could not load face cascade")
    sys.exit(1)

# Initialize MediaPipe Hands
mp_hands = mp.solutions.hands
hands = mp_hands.Hands(max_num_hands=1, min_detection_confidence=0.6, min_tracking_confidence=0.6)
mp_draw = mp.solutions.drawing_utils

# Initialize Webcam
cap = cv2.VideoCapture(0)
if not cap.isOpened():
    print("Error: Could not open webcam. Trying camera index 1...")
    cap = cv2.VideoCapture(1)
    if not cap.isOpened():
        print("Error: Could not open any webcam")
        sys.exit(1)

# Set webcam resolution
cap.set(cv2.CAP_PROP_FRAME_WIDTH, 640)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 480)

# Create a window for the webcam feed
cv2.namedWindow('Face and Gesture Control', cv2.WINDOW_NORMAL)

# Playback and gesture states
is_playing = False
last_gesture = None
gesture_cooldown = 0.3  # Increased sensitivity
last_gesture_time = 0
finger_counts = deque(maxlen=5)
consistent_frames = 0
last_finger_count = 0

try:
    # Start playback
    if player.play() == -1:
        print("Error: Could not start playback")
        sys.exit(1)
    
    time.sleep(0.5)  # Wait for VLC to initialize
    
    # Get video duration
    while player.get_length() <= 0:
        time.sleep(0.1)
    video_duration_ms = player.get_length()
    print(f"Video duration: {video_duration_ms}ms")

    while True:
        ret, frame = cap.read()
        if not ret:
            print("Error: Could not read frame from webcam")
            break
            
        # Flip frame for natural view
        frame = cv2.flip(frame, 1)
        
        # Face detection
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        faces = face_cascade.detectMultiScale(gray, scaleFactor=1.05, minNeighbors=3, minSize=(30, 30))
        
        # Draw face rectangles
        for (x, y, w, h) in faces:
            cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2)
        
        # Print face detection status
        print(f"Detected {len(faces)} faces")

        # Hand gesture detection
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        results = hands.process(frame_rgb)
        
        gesture = None
        raised_fingers = 0
        confidence = 0.0  # Initialize confidence
        if results.multi_hand_landmarks and len(faces) > 0:
            for hand_landmarks in results.multi_hand_landmarks:
                mp_draw.draw_landmarks(frame, hand_landmarks, mp_hands.HAND_CONNECTIONS)
                
                landmarks = hand_landmarks.landmark
                wrist = landmarks[0]
                palm_base = landmarks[5]
                index_tip = landmarks[8]
                middle_tip = landmarks[12]
                ring_tip = landmarks[16]
                pinky_tip = landmarks[20]
                
                hand_size = ((wrist.x - palm_base.x)**2 + (wrist.y - palm_base.y)**2)**0.5
                if hand_size < 0.1:
                    print("Hand too small, ignoring")
                    continue
                
                y_threshold = hand_size * 0.45
                x_separation = hand_size * 0.1
                
                finger_tips = [index_tip, middle_tip, ring_tip, pinky_tip]
                raised_fingers = 0
                for i, tip in enumerate(finger_tips):
                    tip_to_palm_dist = ((tip.x - palm_base.x)**2 + (tip.y - palm_base.y)**2)**0.5
                    if (tip.y < palm_base.y - y_threshold and
                        abs(tip.x - palm_base.x) > x_separation and
                        tip_to_palm_dist > hand_size * 0.4):
                        raised_fingers += 1
                        h, w, _ = frame.shape
                        tip_pos = (int(tip.x * w), int(tip.y * h))
                        cv2.circle(frame, tip_pos, 5, (0, 255, 0), -1)
                        cv2.line(frame, (int(wrist.x * w), int(wrist.y * h)), tip_pos, (0, 255, 0), 3)
                    else:
                        h, w, _ = frame.shape
                        tip_pos = (int(tip.x * w), int(tip.y * h))
                        cv2.circle(frame, tip_pos, 5, (0, 0, 255), -1)
                
                finger_counts.append(raised_fingers)
                if len(finger_counts) == finger_counts.maxlen:
                    raised_fingers = max(set(finger_counts), key=list(finger_counts).count)
                    confidence = list(finger_counts).count(raised_fingers) / len(finger_counts)
                
                if raised_fingers == last_finger_count:
                    consistent_frames += 1
                else:
                    consistent_frames = 0
                    last_finger_count = raised_fingers
                
                current_time = time.time()
                if (current_time - last_gesture_time > gesture_cooldown and
                    consistent_frames >= 1 and
                    confidence > 0.4):
                    if raised_fingers == 1:
                        gesture = "one_finger"
                    elif raised_fingers == 2:
                        gesture = "two_fingers"
                    elif raised_fingers == 3:
                        gesture = "three_fingers"
                    elif raised_fingers == 4:
                        gesture = "four_fingers"
                
                print(f"Hand detected. Raised fingers: {raised_fingers}, Confidence: {confidence:.2f}, "
                      f"Consistent frames: {consistent_frames}, Gesture: {gesture or 'None'}")
        
        # Save debug frame if no gesture
        if not gesture and raised_fingers > 0:
            cv2.imwrite(f"debug_gesture_{int(time.time())}.jpg", frame)

        # Control playback based on face detection
        player_state = player.get_state()
        if len(faces) > 0:
            if player_state == vlc.State.Ended:
                print("Player ended, restarting media")
                player.stop()
                player.set_media(Media)
                player.play()
                time.sleep(0.5)
                is_playing = True
                print("Playing ▶️ (Restarted)")
            elif player_state != vlc.State.Playing:
                print("Attempting to play. Current state:", player_state)
                if player.play() == -1:
                    print("Error: Failed to play")
                else:
                    is_playing = True
                    print("Playing ▶️ (Face Detected)")
        else:
            if player_state == vlc.State.Playing:
                print("Attempting to pause. Current state:", player_state)
                player.pause()
                time.sleep(0.1)
                is_playing = False
                print("Paused ⏸️ (No Face)")

        # Control volume and seeking based on gestures
        if gesture and current_time - last_gesture_time > gesture_cooldown:
            print(f"Gesture triggered: {gesture}, VLC state: {player.get_state()}, "
                  f"Current time: {player.get_time()}ms, Duration: {video_duration_ms}ms")
            if player.get_state() in [vlc.State.Playing, vlc.State.Paused]:
                if gesture == "one_finger":
                    current_volume = player.audio_get_volume()
                    if current_volume < 100:
                        new_volume = min(current_volume + 10, 100)
                        result = player.audio_set_volume(new_volume)
                        time.sleep(0.1)
                        print(f"Volume Up 📈: Attempted {new_volume}%, Result: {result}, Current: {player.audio_get_volume()}%")
                    else:
                        print("Volume Up 📈: Already at max (100%)")
                        cv2.putText(frame, "Volume Max", (10, 120), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)
                elif gesture == "two_fingers":
                    current_volume = player.audio_get_volume()
                    if current_volume > 0:
                        new_volume = max(current_volume - 10, 0)
                        result = player.audio_set_volume(new_volume)
                        time.sleep(0.1)
                        print(f"Volume Down 📉: Attempted {new_volume}%, Result: {result}, Current: {player.audio_get_volume()}%")
                    else:
                        print("Volume Down 📉: Already at min (0%)")
                        cv2.putText(frame, "Volume Min", (10, 120), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)
                elif gesture == "three_fingers":
                    current_time_ms = player.get_time()
                    new_time_ms = current_time_ms + 3000
                    if new_time_ms < video_duration_ms:
                        player.set_time(new_time_ms)
                        time.sleep(0.1)
                        print(f"Fast Forward 3s ⏩: Attempted {new_time_ms}ms, Current: {player.get_time()}ms")
                    else:
                        print("Fast Forward 3s ⏩: Cannot seek past video end")
                        cv2.putText(frame, "End of Video", (10, 120), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)
                elif gesture == "four_fingers":
                    current_time_ms = player.get_time()
                    new_time_ms = max(current_time_ms - 3000, 0)
                    player.set_time(new_time_ms)
                    time.sleep(0.1)
                    print(f"Rewind 3s ⏪: Attempted {new_time_ms}ms, Current: {player.get_time()}ms")
            
            else:
                print(f"Gesture {gesture} ignored: Invalid VLC state {player.get_state()}")
                cv2.putText(frame, f"Invalid VLC State: {player.get_state()}", (10, 120), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)
            
            last_gesture_time = current_time
        elif gesture:
            print(f"Gesture {gesture} not triggered: Cooldown={current_time - last_gesture_time:.2f}s, "
                  f"Consistent={consistent_frames}, Confidence={confidence:.2f}")

        last_gesture = gesture

        # Display status on frame
        status = "Playing" if is_playing else "Paused"
        cv2.putText(frame, f"Status: {status}", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
        cv2.putText(frame, f"Gesture: {gesture or 'None'}", (10, 60), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
        cv2.putText(frame, f"Fingers: {raised_fingers}", (10, 90), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

        cv2.imshow('Face and Gesture Control', frame)

        # Press 'q' to quit
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

finally:
    cap.release()
    cv2.destroyAllWindows()
    player.stop()
    hands.close()
```
